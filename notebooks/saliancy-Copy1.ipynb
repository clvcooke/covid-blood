```python
def imshow(img, transpose = True):
    img = img*0.229 + 0.485    # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
    
def attribute_image_features(algorithm, inp,net, labels, ind, **kwargs):
    net.zero_grad()
    tensor_attributions = algorithm.attribute(inp,
                                              target=labels[ind],
                                              **kwargs
                                             )
    
    return tensor_attributions
```
