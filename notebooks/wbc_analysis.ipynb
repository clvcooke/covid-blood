```python
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
from tqdm import tqdm_notebook as tqdm
import tensorflow.keras as keras
import matplotlib
```

```python
def read_img(path, channel=None):
    if channel is None:
        return cv.cvtColor(cv.imread(path), cv.COLOR_BGR2RGB)
    elif channel == 'h':
        return cv.cvtColor(cv.imread(path), cv.COLOR_BGR2HSV)[:,:,0]
    elif channel == 's':
        return cv.cvtColor(cv.imread(path), cv.COLOR_BGR2HSV)[:,:,1]
    elif channel == 'v':
        return cv.cvtColor(cv.imread(path), cv.COLOR_BGR2HSV)[:,:,2]
```
