```python
with open('/home/colin/covid-blood/notebooks/prep/outcome_by_order.json') as fp:
    outcome_data = json.load(fp)
with open('/home/colin/cbc_data.json') as fp:
    cbc_data = json.load(fp)
```

```python
platelet_counts = []
positive_counts = []
negative_counts = []
for order, cbc in cbc_data.items():
    if 'PLATELET COUNT     (BKR)' in cbc:
        count = cbc['PLATELET COUNT     (BKR)']
        if not np.isnan(count) and order in all_images:
            platelet_counts.append(count)
            if order in positive_images:
                positive_counts.append(count)
            else:
                negative_counts.append(count)
    else:
        print("nope")
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
ax = axs[0]
_, edges, _ = ax.hist(platelet_counts, linestyle='dashed', edgecolor='black')
ax.set_xlabel("Platelet Count")
ax.set_ylabel("Patient Frequency")
ax.set_title("All Patients: Platelet Count Distribution")

ax = axs[1]
amnts_pos, edges_pos, _ = ax.hist(positive_counts, linestyle='dashed', edgecolor='black', bins=edges)
ax.set_xlabel("Platelet Count")
ax.set_ylabel("Patient Frequency")
ax.set_title("Positive Patients: Platelet Count Distribution")

ax = axs[2]
amnts_neg, edges_neg, _ = ax.hist(negative_counts, linestyle='dashed', edgecolor='black', bins=edges)
ax.set_xlabel("Platelet Count")
ax.set_ylabel("Patient Frequency")
ax.set_title("Negative Patients: Platelet Count Distribution")
```
