```python
def disp_images(images):
    fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(6,6))
    for i in range(6):
        x = int(i // 2)
        y = int(i % 2)
        ax = axs[x,y]
        ax.imshow(images[i])
        ax.axis('off')
```

```python
def filter_wbc(image, lower_bound = (0,100,0), upper_bound = (204, 255, 127)):
    image_g = image.copy()
    image_s = cv.cvtColor(image_g, cv.COLOR_RGB2LAB)
    mask = cv.inRange(image_s, lower_bound, upper_bound)
    mask = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel=np.ones((9,9), np.uint8), iterations=1)
    mask_rgb = cv.cvtColor(mask,cv.COLOR_GRAY2RGB)
    image_g = image_g & ~mask_rgb
    return image_g
disp_images([filter_wbc(im) for im in test_images])
disp_images([im for im in test_images])
plt.show()
```
